# Data analysis of 82 Stripe Data set 
### the data set includes stars and quasars' location, 5 bands' magnitudes and redshift.

```python
import numpy as np
import pandas as pd
```

```python
data = pd.read_csv('../../../data/processed/unbalanced/final_v1.csv')
```

### Main Sequence


```python
id_list = []
combine = data.groupby(data['id']) 
for _id, group in combine:
    id_list.append(_id)
u_g = []
g_r = []
z = []
label = []
for i in id_list:
	ind_list = data[data.id == i].index.tolist()
	g_mean = data['g'][ind_list].tolist()[0]
	r_mean = data['r'][ind_list].tolist()[0]
	u_mean = data['u'][ind_list].tolist()[0]
# 	g_mean = np.mean(data['g'][ind_list].tolist())
# 	r_mean = np.mean(data['r'][ind_list].tolist())
# 	u_mean = np.mean(data['u'][ind_list].tolist())
	g_r.append(g_mean-r_mean)
	u_g.append(u_mean-g_mean)
	z.append(data['redshift'][ind_list[0]])
	label.append(data['type'][ind_list[0]])
    

    
```

```python
print(np.min(z))
print(np.max(z))
```

```python
import matplotlib.pyplot as plt
from matplotlib.pylab import *
plt.figure(figsize = (9,11))
cm = plt.cm.get_cmap('RdYlBu')
img = plt.scatter(u_g, g_r,c=z, vmin=np.min(z), vmax=3, s=1, cmap=cm)
plt.xlabel('(u - g) color', fontsize=16)
plt.ylabel('(g - r) color', fontsize=16)
fontsize        = 16
labelsize       = fontsize
tickwidth       = 2.0
ticklength      = 2.0
cb = plt.colorbar(img,orientation="horizontal").set_label(label='redshift',size=16)

plt.tick_params(labelsize=labelsize, length=ticklength, width=tickwidth)
plt.xlim(-4,7) 
plt.ylim(-4,7)

plt.savefig('plots/distribution.png')
```

```python
id_list = []
combine = data.groupby(data['id']) 
for _id, group in combine:
    id_list.append(_id)
u_g_qso = []
g_r_qso = []
u_g_star = []
g_r_star = []
z = []
label = []
for i in id_list:
	ind_list = data[data.id == i].index.tolist()       
	g_mean = np.mean(data['g'][ind_list].tolist())
	r_mean = np.mean(data['r'][ind_list].tolist())
	u_mean = np.mean(data['u'][ind_list].tolist())
	if data['type'][ind_list[0]]==1:
		g_r_qso.append(g_mean-r_mean)
		u_g_qso.append(u_mean-g_mean)
	else:
		g_r_star.append(g_mean-r_mean)
		u_g_star.append(u_mean-g_mean)      
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize = (9,9))
plt.scatter(u_g_qso, g_r_qso,color='b', s=1,label='Quasar')
plt.scatter(u_g_star, g_r_star, color='r', s=1,label='non-Quasar')
plt.legend(loc='upper right',fontsize=16)
plt.xlabel('(u - g) color', fontsize=16)
plt.ylabel('(g - r) color', fontsize=16)
plt.xlim(-4,7) 
plt.ylim(-4,7)
plt.savefig('distribution_class.png')
```

### Matrix plot of 40,000 Stripe 82 point source objects over the span of 10 years

```python
mjd_list = data['mjd']
mjd_max = np.nanmax(mjd_list)
max_obj = 4000

print("the time range is from %d to %d (day)"% ((mjd_max-8*356),mjd_max))
use_mjd = [x for x in range((int(mjd_max)-8*356),int(mjd_max))]

# pick max_obj number of objects and sort them by their redshift
obj = None
n = 0
num_obj =0
z_dict = {}
while n < len(data) and num_obj <= max_obj:
    if data['id'][n]!=obj:
        obj=data['id'][n]
        z_dict[obj]=data['redshift'][n]
        num_obj +=1
    n +=1
z_order = sorted(z_dict.items(), key=lambda x:x[1])
obj_list =[x[0] for x in z_order]

#convert to the obj id in the plot for convenience
map_obj = {}
n = 0
for i in obj_list:
    map_obj[i] = n
    n +=1
    
```

```python
n = 0
obj = None
data_mag = []

df = []
for i in obj_list:
    mag_list = []
    ind_list = data[data.id == i].index.tolist()
    mjd_exit_list = data['mjd'][ind_list].tolist()
    mjd_exit_list = [int(x) for x in mjd_exit_list]
    
    m = 0
    while m<len(use_mjd):
        if use_mjd[m] in mjd_exit_list:
            mag = data['g'][ind_list[mjd_exit_list.index(use_mjd[m])]]
            mag_list.append(mag)
        else:
            mag_list.append(np.nan)
        m +=1
    df.append(mag_list)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
# df2 = pd.DataFrame(np.array(df))
test = np.array(df)
print(test.shape)
f, ax = plt.subplots(figsize=(32, 16))
# plt.figure(figsize = (32,16))
ax = sns.heatmap(test, cmap='Reds_r',vmin=15,vmax=25, label='g magnitude')
plt.xticks(fontsize=20)
plt.yticks(fontsize=20)
# sns.set_label('g magnitude',fontsize=24)
cax = plt.gcf().axes[-1]
cax.tick_params(labelsize=20)
#设置colorbar的label文本和字体大小
cbar = ax.collections[0].colorbar
cbar.set_label('g magnitude',fontsize=24)

locs, labels = plt.xticks()
print(locs)
print(labels)
x_tick = np.linspace(0,2848,50)
x_label = [str(int(x)) for x in x_tick]
print(x_tick)
print(x_label)
plt.xticks(x_tick, x_label, fontsize=20)
plt.yticks([])

plt.xlabel('days from earliest observation', fontsize=24, labelpad=10)
plt.ylabel('object sorted by redshift', fontsize=24)
plt.savefig('plots/heatmap8years_u.png')
```

### Observation gap time, error, length analysis

##### delta_time: The time difference between two neighboring observations
##### data length: the number of detections for one object
##### observation length: The time difference between the earliest and the latest observations

```python
combine = data.groupby(data['id']) 

# create the id list
id_list = []
for _id, group in combine:
    id_list.append(_id)

delta_time = []
observation_time = []
redshift = []
for i in id_list:
    ind_list = data[data.id == i].index.tolist()
    mjd_exit_list = data['mjd'][ind_list].tolist()
    redshift.append(data['redshift'][ind_list].tolist()[0])
    m = 0
    while m<len(mjd_exit_list)-1:
        delta_time.append(mjd_exit_list[m+1]-mjd_exit_list[m])
        m +=1
    observation_time.append(mjd_exit_list[-1]-mjd_exit_list[0])
    
    
```

```python
import matplotlib.pyplot as plt
# plt.figure(figsize=(10, 5))
plt.figure(figsize = (8,8))
# plt.subplot(1,2,1)
plt.hist(x=delta_time,bins=10**np.arange(0,3,0.1))# bins=10**np.arange(0,3,0.1)
plt.xlabel(r'$\Delta$ t (days)', fontsize=20)
plt.ylabel('Number', fontsize=20)
plt.xscale('log')
plt.savefig('delta_time.png')

# plt.subplot(1,2,2)
plt.figure(figsize = (8,8))
obj_list = combine['id'].count().tolist()
plt.hist(x=obj_list)
plt.xlabel('data length', fontsize=20)
plt.ylabel('Number', fontsize=20)
plt.savefig('data_length.png')
# plt.tight_layout()
print(me)
```
